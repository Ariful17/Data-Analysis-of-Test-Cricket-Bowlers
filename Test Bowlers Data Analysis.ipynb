**Renaming the Column Appropriately**

```python
df=df.rename(columns={'Span':'Career Length',
                     'Mat':'Matches',
                     'Inns':'Innings',
                     'Balls':'Total Balls in Career',
                     'Runs':'Total Runs Given in Career',
                     'Wkts':'Total Wickets in Career',
                     'BBI':'Best Bowling in Innings',
                     'BBM':'Best Bowling in Matches(Two Innings in a Test Match)',
                     'Ave':'Average(Runs per Wicket)',
                     'Econ':'Economy Rate(Average Runs per Over)',
                     'SR':'Bowling Strike Rate(Average Number of Balls per Wicket)',
                     '5':'Number of times 5 Wickets taken in an Innings',
                     '10':'Number of times 10 wickets taken in an Innings'})
display(df)
```

**Removing a Column**

```python
df=df.drop('Total Balls in Career', axis=1)
display(df)
```

The column showing total balls in career has been removed
